```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from config import key
from citipy import citipy
```

```python
lat_range = (-90, 90)
lng_range = (-180, 180)

lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

query_url = f"{url}appid={key}&units={units}&q="

test = requests.get(query_url + cities[0]).json()
test
```

```python
city = []
data = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

for x in cities:
    try:
        response = requests.get(query_url + x).json()
        data.append(response)
        city.append(response['name'])
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed'])
        print(f"Data gathered for {x}.")
    except:
        print(f"Data not found for {x}. Trying next city....")
```

```python
weather_labels = {
    "City": city,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Latitude": lat,
    "Longitude": lng,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed
}
weather_df = pd.DataFrame(weather_labels)
weather_df.to_csv("output_data.csv")
weather_df.count()
```
